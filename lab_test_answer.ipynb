```python

def generate_personalized_diet(student):
    if student["weight"] <= 50:
        return "diet with high protien and lots of fruits and vegetables."
    elif 50 < student["weight"] <= 70:
        return "moderate protiens and carb content."
    else:
        return "high fiber with low fat content."


def recommend_sport(student):
    if student["agility"] >= 8 and student["speed"] >= 8:
        return "Football"
    elif student["strength"] >= 9:
        return "Javelin Throw"
    elif student["agility"] >= 7 and student["speed"] >= 7:
        return "Cricket"
    else:
        return "Chess"
    
    

def display_info():
    for student in students_info:
        print(f"Name: {student['name']}")
        print(f"Class: {student['class']}")
        print(f"Height: {student['height']} cm")
        print(f"Weight: {student['weight']} kg")
        print(f"Preferred Sport: {student['preferred_sport']}")
        diet_plan = generate_personalized_diet(student)
        recommended_sport = recommend_sport(student)
        print(f"Personalized Diet Plan: {diet_plan}")
        print(f"Recommended Sport(based on agility, strength, speed): {recommended_sport}")
        print("\n")

```

```python
display_info()
```
